```python
import nltk
from nltk.stem import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
words = ['charging' , 'charges', 'charged', 'study', 'studying', 'easliy', 'fairly','losses', 'was' , 'neccessary ', 'implementation ', 'permission' , 'learn']
```

```python
for word in words:
  print(word + ' : ' + ps.stem(word))
```

```python
segment = "police is charging the support a sit ran away".split()
```

```python
segment
```

```python
for word in segment:
  print(word + ' : ' + ps.stem(word))
```

```python
from nltk.stem.snowball import SnowballStemmer
snowball = SnowballStemmer(language='english')
```

```python
for word in words:
    print(f"{word} : {snowball.stem(word)} ")
```
