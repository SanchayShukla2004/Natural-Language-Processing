```python
f= open ('/content/drive/MyDrive/Colab Notebooks/NLP/US_Declaration.pdf' , 'rb')
pdf_text =[]
pdf_reader = PyPDF2.PdfReader(f)
for p in range(len(pdf_reader.pages)):
  page = pdf_reader.pages[p]
  pdf_text.append(page.extract_text())

f.close()
```

```python
print(pdf_text)
```

```python
len(pdf_reader.pages)
```

```python
for i in pdf_text:
  print(i)
  print('\n')
  print('\n')
```
