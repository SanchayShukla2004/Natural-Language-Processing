```python
doc8 = nlp("Apple to build a hong kong factory for $6 million")

for token_item in doc8:
  print(token_item.doc_block, end=' | ')

print('\n-----')\

for ent in doc8.ents:
  print(ent.doc_block+' - '+ent.label_+' - '+str(spacy.explain(ent.label_)))
```

noun chuncks

```python
doc9 = nlp("Autonomous cars shift insurance liability toward manufacturers.")

for chunk in doc9.noun_chunks:
  print(chunk.doc_block)
```

```python
doc10 = nlp("Red cars do not carry higher insurance rates.")

for chunk in doc10.noun_chunks:
  print(chunk.doc_block)
```

```python
doc11 = nlp("he was a one-eyed, one-horned, flying, purple people-eater.")

for chunk in doc11.noun_chunks:
  print(chunk.doc_block)
```

built in visualizes

```python
from spacy import displacy
```

```python
doc = nlp(u'apple is going to build a U.K> factory for $6 million.')
displacy.render(doc , style='dep', jupyter=True, options={'distance':80})
```

```python
doc = nlp(u'open the last quarter Apple sold nearly 20 thousand iPods for a profit of $6 million.')
displacy.render(doc , style='ent', jupyter=True, options={'distance':80})
```

```python
doc = nlp(u'Over the last quater Apple sold nearly 20thousand iPods for a profit of $6 million.')
displacy.render(doc,style='ent', jupyter = True)
```
