```python
import nltk
import pandas as pd
import PyPDF2
import spacy
from nltk.tokenize import word_tokenize, sent_tokenize
```

```python
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('punkt_tab')
nltk.download('averaged_perceptron_tagger_eng')
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
pdf_file = open(list(uploaded.keys())[0], 'rb')
reader = PyPDF2.PdfReader(pdf_file)

text = ""
for page in reader.pages:
    text += page.extract_text()

text[:1000]
```

```python
sentences = [s.strip() for s in sent_tokenize(text) if len(s.strip()) > 30]
df = pd.DataFrame(sentences, columns=["Sentence"])
df.head()
```

```python
def pos_tag_sentence(sentence):
    tokens = word_tokenize(sentence)
    return nltk.pos_tag(tokens)
```

```python
df["Tokens"] = df["Sentence"].apply(word_tokenize)
df["POS_Tags"] = df["Sentence"].apply(pos_tag_sentence)

df.head()
```

```python
pos_tags = []
for tags in df["POS_Tags"]:
    for word, tag in tags:
        pos_tags.append(tag)

pos_freq = pd.Series(pos_tags).value_counts()
pos_freq.head(10)
```
